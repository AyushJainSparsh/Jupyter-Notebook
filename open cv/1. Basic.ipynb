```python
plt.imshow(image_hsv)
```

```python
hue , sat , val =cv2.split(image_hsv)
```

```python
val
```

```python
import numpy as np
```

```python
new_val=val.astype(np.int64)+50
new_val
```

```python
new_val[new_val>255]=255
new_val_uint8=new_val.astype(np.uint8)
```

```python
new_val_uint8
```

```python
image_bright_hsv=cv2.merge([hue,sat,new_val_uint8])
```

```python
image_bright_rgb=cv2.cvtColor(image_bright_hsv,cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(image_bright_rgb)
```
