### Perform Data Visualization and perform following graph using matplotlib library

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
days = [1,2,3,4,5,6,7,8,9]
temp = [27,29,31,25,30,32,31,28,28]
```

# Line Graph

```python
plt.plot(days , temp , color = 'b' , marker = 'o' , linestyle = '-.' , label = 'Temp')
plt.title("D vs T")
plt.xlabel("Days")
plt.ylabel('Temperature')
plt.legend()
```

```python
math = np.random.randint(0,100,(100))
chem = np.random.randint(0,100,(100))
print(math)
print(chem)
```

# Histogram

```python
plt.hist([math , chem] , color = ['y','b'] , label = ['Math' , 'Chem'])
plt.title('Student Marks Frequency')
plt.xlabel('Marks')
plt.legend()
```

# Scatter Plot

```python
plt.scatter(days , temp , color = 'orange' , marker = 'x' , label = 'Temp')
plt.title("D vs T")
plt.xlabel("Days")
plt.ylabel('Temperature')
plt.legend()
```

# Bar Graph

```python
plt.bar(days , temp , color = 'g'  , label = 'Temp')
plt.title("D vs T")
plt.xlabel("Days")
plt.ylabel('Temperature')
plt.legend()
```

# Pie Chart

```python
cars = ['AUDI', 'BMW', 'FORD','TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]
fig = plt.figure(figsize=(10, 7))
plt.pie(data , labels=cars)
plt.legend()
plt.show()
```

# Box Plot

```python
plt.boxplot(temp , patch_artist=True, vert=False)
plt.title("Temperature")
```

# Heat Map

```python
data = np.random.random(( 10 , 10 )) 
plt.imshow( data , vmin = -1.0 , vmax = 1.0  )
plt.colorbar()
plt.title( "2-D Heat Map" ) 
plt.show() 
```
