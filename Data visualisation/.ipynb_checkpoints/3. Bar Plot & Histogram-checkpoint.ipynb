```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
churn = pd.read_csv("Churn_Modelling.csv")
```

```python
churn
```

```python
churn.info()
```

```python
churn["Gender"].unique()
```

```python
churn_gender_es=churn.groupby("Gender")[["EstimatedSalary","CreditScore"]].mean().reset_index()
```

```python
churn_gender_es
```

```python
churn_gender_es.plot.bar(x="Gender",y="EstimatedSalary")
```

```python
churn_gender_es.plot.bar(x="Gender")
```

```python
plt.bar(churn_gender_es["Gender"],churn_gender_es["EstimatedSalary"])
```

```python
sns.barplot(churn_gender_es , x="Gender" , y= "EstimatedSalary")
```

```python
sns.barplot(data= churn_gender_es , y="Gender" , x= "EstimatedSalary" , orient="h")
```

```python
churn_gender_es.plot.barh(x="Gender",y="EstimatedSalary")
```

```python
plt.barh(churn_gender_es["Gender"],churn_gender_es["EstimatedSalary"])
```
